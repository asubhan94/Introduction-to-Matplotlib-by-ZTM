# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [0, 1, 2, 3, 4]
y = [11, 22, 33, 44, 55]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 10)) #(width, height)

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple plot', xlabel='x-axis', ylabel='y-axis')

# 5. Save and show (you save the figure)
fig.savefig('images/sample-plot.png')
```

### Two options for subplots

```python
# subplots option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# subplots option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib OO method?)

* when plotting something quickly, okay to use the pyplot method (just to visualize data)
* when plotting something advance, use the OO method

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method

over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

# ax.set_xlim([45, 100])
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);

# Customize the plot
ax.set(title='heart disease and cholestrol levels',
       xlabel='Age',
       ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# subplots of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])

# customize ax0
ax0.set(title='heart disease and cholestrol level',
        ylabel='cholestrol')

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add a mean line
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

# Add data to ax0
scatter1 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'])
# customize ax0
ax1.set(title='heart disease and heart beat rate',
        xlabel='Age',
        ylabel='thalach')
# add a legend to ax0
ax1.legend(*scatter.legend_elements(), title='Target')

# add a mean line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# Add a title to figure
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');
```

```python
# set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter') # this change the color scheme

# Customize the plot
ax.set(title='heart disease and cholestrol levels',
       xlabel='Age',
       ylabel='Cholestrol')

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),

           linestyle='--');
```

```python
# cutomising the y and x axis limitation

# subplots of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')

# customize ax0
ax0.set(title='heart disease and cholestrol level',
        ylabel='cholestrol')

# change the x axis limits
ax0.set_xlim([50, 80])

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add a mean line
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

# Add data to ax0
scatter1 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'],
                       cmap='winter')
# customize ax0
ax1.set(title='heart disease and heart beat rate',
        xlabel='Age',
        ylabel='thalach')

# change the ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# add a legend to ax0
ax1.legend(*scatter.legend_elements(), title='Target')

# add a mean line
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--');

# Add a title to figure
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');

```
